```python
import requests 
import pandas as pd
import numpy as np 
from bs4 import BeautifulSoup
from pandas import DataFrame as df  
```

```python
############################################ Creating the URL dataframe ######################################################
URL_list = []
page_quantity = (10)
# URL = 'https://ca.indeed.com/salaries/Data-Scientist-Salaries' + '?start=' + page_number

# print (URL)

for i in range( page_quantity ):
    page_number = str(i*10)
    page_number
    URL = 'https://ca.indeed.com/salaries/Data-Scientist-Salaries' + '?start=' + page_number
    URL_list.append(URL)
URL_list

URL_df = pd.DataFrame (URL_list)
URL_df.columns = ['URL']
# URL_df.set_index('URL', inplace=True, drop=True)
URL_df
```

```python
############################################   Creating the empty list   ######################################################
salary_list = []


####################################    Loop on the URLs and creating the Dataframe      ###################################### 
URL_column = URL_df.loc[: , 'URL']
# print (URL_column)
for i in URL_column:
    
    r = requests.get (i)
    soup = BeautifulSoup (r.text, 'html.parser')
    
    for i in soup.find_all('table', attrs = { 'id' : "cmp-sal-company-table"}):
    #     print (i)
        for s in i:
            string = soup.find('tbody')
        for c in string:
    #         print (c.text)
            companies = c.text

            Data_index = companies.find ('Data')
            Salary_index = companies.find ('$')    
            company = companies[:Data_index]
            salary = companies[Salary_index:]
            string = company , salary
            salary_list.append(string)

        
# print (salary_list)
############################################ Creating the final dataframe ######################################################
dataframe_salaries = pd.DataFrame (salary_list)
dataframe_salaries.columns = ['Company', 'Salary']
dataframe_salaries.set_index('Company', inplace=True, drop=True)
dataframe_salaries
```
